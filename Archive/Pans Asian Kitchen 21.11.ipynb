```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

# Function to scroll within a specific panel
def scroll_panel_with_page_down(driver, panel_xpath, presses, pause_time):
    panel_element = driver.find_element(By.XPATH, panel_xpath)
    actions = ActionChains(driver)
    actions.move_to_element(panel_element).click().perform()

    for _ in range(presses):
        actions = ActionChains(driver)
        actions.send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(pause_time)

# Function to click "Mehr" buttons
def click_more_buttons(driver):
    try:
        more_buttons = driver.find_elements(By.XPATH, './/button[@aria-label="Mehr"]')
        for button in more_buttons:
            try:
                button.click()
                time.sleep(0.5)
            except Exception as e:
                print(f"Could not click 'Mehr' button: {e}")
    except Exception as e:
        print(f"Error finding 'Mehr' buttons: {e}")

# Function to extract usernames and raw review text
def extract_reviews(driver, username_selector, reviews_selector):
    wait = WebDriverWait(driver, 10)
    review_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, reviews_selector)))
    reviewer_names = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, username_selector)))

    data = []
    for name_elem, review_elem in zip(reviewer_names, review_elements):
        data.append({
            "Username": name_elem.text.strip(),
            "Review": review_elem.get_attribute('innerText')  # Captures raw text
        })
    return data

# Regex-based category extraction functions
def extract_initial_review_text(text):
    """Capture the initial review text up to the first detected category."""
    pattern = r"^(.*?)(?=\n(?:Servicetyp|Art der Mahlzeit|Preis pro Person|Essen:|Service:|Ambiente:|Empfehlungen für Gerichte|Empfehlung für Vegetarier|Vegetarische Angebote|$))"
    match = re.search(pattern, text, re.DOTALL)
    return match.group(1).strip() if match else text.strip()

def extract_category_content(text, category):
    """
    Extract content for a specific category.
    
    :param text: Full review text.
    :param category: The category keyword to extract content for.
    :return: Extracted content or 'none' if not found.
    """
    pattern = rf"{category}\s*(.*?)(?=\n(?:Servicetyp|Art der Mahlzeit|Preis pro Person|Essen:|Service:|Ambiente:|Empfehlungen für Gerichte|Empfehlung für Vegetarier|Vegetarische Angebote|Vegetarische Gerichte|Ernährungsbeschränkungen|Parkplätze|Parkplatz|Kinderfreundlichkeit|Zugänglichkeit m.d. Rollstuhl|Ende|$))"
    match = re.search(pattern, text, re.DOTALL)
    return match.group(1).strip() if match else "none"

# Define the XPath to the scrollable panel and CSS selectors
panel_xpath = "//*[@id='QA0Szd']/div/div/div[1]/div[2]/div"
username_selector = ".d4r55"
reviews_selector = "MyEned"

# Scroll and click "Mehr" buttons to reveal full reviews
scroll_panel_with_page_down(driver, panel_xpath, presses=100, pause_time=1)
click_more_buttons(driver)

# Extract reviewer names and reviews
reviews_data = extract_reviews(driver, username_selector, reviews_selector)

# Parse and categorize reviews
parsed_reviews = []
for entry in reviews_data:
    review_txt = entry["Review"]

    # Extract initial review text and categories
    initial_review_text = extract_initial_review_text(review_txt)
    servicetyp_text = extract_category_content(review_txt, r"Servicetyp")
    artdermahlzeit_text = extract_category_content(review_txt, r"Art der Mahlzeit")
    preisproperson_text = extract_category_content(review_txt, r"Preis pro Person")
    essen_text = extract_category_content(review_txt, r"Essen:")
    service_text = extract_category_content(review_txt, r"Service:")
    ambiente_text = extract_category_content(review_txt, r"Ambiente:")
    empfehlungenfuergerichte_text = extract_category_content(review_txt, r"Empfehlungen für Gerichte")
    empfehlungfuervegetarier_text = extract_category_content(review_txt, r"Empfehlung für Vegetarier")
    vegetarischeangebote_text = extract_category_content(review_txt, "Vegetarische Angebote")
    vegetarischegerichte_text = extract_category_content(review_txt, "Vegetarische Gerichte")
    ernaehrungseinschraenkungen_text = extract_category_content(review_txt, "Ernährungseinschränkungen")
    artderparkplaetze_text = extract_category_content(review_txt, "Art der Parkplätze")
    parkplatz_text = extract_category_content(review_txt, "Parkplatz")
    kinderfreundlichkeit_text = extract_category_content(review_txt, "Kinderfreundlichkeit")
    zugaenglichkeitmdrollstuhl_text = extract_category_content(review_txt, "Zugänglichkeit m.d. Rollstuhl")

    # Append parsed data
    parsed_reviews.append({
        "Username": entry["Username"],
        "Review": initial_review_text,
        "Servicetyp": servicetyp_text,
        "Art der Mahlzeit": artdermahlzeit_text,
        "Preis pro Person": preisproperson_text,
        "Essen": essen_text,
        "Service": service_text,
        "Ambiente": ambiente_text,
        "Empfehlungen für Gerichte": empfehlungenfuergerichte_text,
        "Empfehlung für Vegetarier": empfehlungfuervegetarier_text,
        "Vegetarische Angebote": vegetarischeangebote_text,
        "Vegetarische Gerichte": vegetarischegerichte_text,
        "Ernährungsbeschränkungen": ernaehrungseinschraenkungen_text,
        "Art der Parkplätze": artderparkplaetze_text,
        "Parkplatz": parkplatz_text,
        "Kinderfreundlichkeit": kinderfreundlichkeit_text,
        "Zugänglichkeit m.d. Rollstuhl": zugaenglichkeitmdrollstuhl_text
    })

# Convert parsed reviews to DataFrame
df = pd.DataFrame(parsed_reviews)

# Save to CSV
csv_file_path = 'C:/Users/Theresa/OneDrive/Master/places_detailed.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8')

print(f"Detailed reviews successfully saved to {csv_file_path}")
```
